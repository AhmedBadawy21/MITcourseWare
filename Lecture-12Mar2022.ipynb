```python
import requests
from bs4 import BeautifulSoup

url = "https://jekso.github.io/scrapping-example/index.html?"

response = requests.get(url)

print(response.text)
```

```python
soup = BeautifulSoup(response.text)
```

```python
print(soup.prettify())
```

```python
p = soup.find('p')

p.get_text()
p.text

all_p = soup.find_all('p')
all_h1 = soup.find_all('h1')

print(all_p[0].text)
print(all_h1)
```

```python
orange_h2 = soup.find_all("h2", attrs={'class':'h2_orange'})

print(orange_h2)
```

```python
yellow_h2 = soup.find_all("h2", class_ ='bg-yellow')
yellow2_h2 = soup.find_all("h2", attrs={'class':'bg-yellow'})

print(yellow2_h2[0].text)
```

```python
yellow_h2[0].has_attr('class')
```

```python
yellow_h2[0].has_attr('href') #href is a link
```
